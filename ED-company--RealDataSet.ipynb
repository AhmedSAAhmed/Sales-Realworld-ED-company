## EDA FOR ED DATASET

- ## What is the overall sales trend

```python
ed.head()
```

```python
ed['order_date'].min()
```

```python
ed['order_date'].max()
```

```python
ed['month_year'] = ed['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
ed['month_year']
```

```python
ed_trend = ed.groupby('month_year').sum()['sales'].reset_index()
```

```python
ed_trend
```

```python
#setting the fig size
plt.figure(figsize=(20,8))
plt.plot(ed_trend['month_year'],ed_trend['sales']  ,color ='red')
plt.xticks(rotation='vertical', size=8)
plt.show()
```

- ## Which are the Top 10 products by sales?

```python
ed.head()
```

```python
prod_sales = pd.DataFrame(ed.groupby('product_name').sum()['sales'])
```

```python
prod_sales
```

```python
prod_sales_desc = prod_sales.sort_values('sales', ascending =False)
```

```python
prod_sales_desc.head()
```

```python
prod_sales_desc[:10]
```

- ## Which are the Most Selling Products?



```python
    #grouping product name
most_selling_prod = ed.groupby('product_name').sum()['quantity']
```

```python
most_selling_prod
```

```python
most_selling_prod = most_selling_prod.sort_values(ascending=False).reset_index()
```

```python
most_selling_prod[:10]
```
